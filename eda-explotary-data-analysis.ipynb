Citizen_ID - A unique alphanumeric identifier for each citizen (e.g., CIT_12345).

Diet_Type - The primary dietary habit of the citizen (e.g., Seafood, Seaweed, Exotic Imports).

District_Name- The name of the district where the citizen currently resides.

Wealth_Index- A numerical score representing the citizen's accumulated wealth and assets. (Note: Contains missing values).

House_Size_sq_ft - The total area of the citizen's living quarters in square feet. (Note: Contains missing values).

Life_Expectancy- The projected lifespan of the citizen based on health and environmental factors.

Vehicle_Owned - The mode of underwater transportation owned by the citizen (e.g., Sea Scooter, Submarine, Fin Bicycle).

Work_District- The district where the citizen travels to for their daily employment.

Bio_Hash - An encrypted string representing the citizen's unique biological signature.

Occupation - [TARGET] The profession of the citizen (Warrior, Merchant, Fisher, Miner, Scribe). This is the column you must predict.

```python
df['Occupation']=df['Occupation'].map({'Warrior': 0, 'Merchant': 1, 'Fisher': 2, 'Miner': 3, 'Scribe': 4})
```
